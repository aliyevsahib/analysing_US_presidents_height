```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data =  pd.read_csv('/content/drive/MyDrive/US_presidents_height/president_heights.csv')
print(data.head(5))
heights = np.array(data['height(cm)'])
print(heights)
```

```python
sorted_data = data.sort_values(by="height(cm)", ascending=False)
top10 = sorted_data.head(10)

print("Top 10 Tallest U.S. Presidents:")
print(top10[["name", "height(cm)"]])
```

```python
print("Mea height:         ", heights.mean())
print("Standard deviation: ", heights.std())
print("Minimum height:     ", heights.min())
print("Maximum height:     ", heights.max())
```

```python
print("25th perceptile:     ", np.percentile(heights, 25))
print("Median:              ", np.median(heights))
print("75th perpercentile:  ", np.percentile(heights, 75))
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-whitegrid')
plt.hist(heights)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height (cm)')
plt.ylabel('number');
```
